```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Titik awal (persegi dekat sumbu X)
square_init = np.array([
    [1,1],
    [3,1],
    [3,2],
    [1,2]
])

dx = 0
dy = 1
steps = 6

fig, ax = plt.subplots(figsize=(6,6))

# Grid & sumbu (lebih simetris)
ax.set_xlim(0, 6)
ax.set_ylim(-8, 8)
ax.set_xticks(range(0,7))
ax.set_yticks(range(-8,9))
ax.grid(True)

ax.axhline(0, linewidth=1.5)  # sumbu X (cermin)
ax.axvline(0, linewidth=1)

original_plot, = ax.plot([], [], 'bo-', label='Objek Asli')
reflection_plot, = ax.plot([], [], 'ro--', label='Refleksi')

ax.legend()
ax.set_title("Refleksi Sumbu X (Menjauh dari Cermin)")

texts = []

def update(frame):
    nonlocal_square = square_init + np.array([dx, dy]) * frame

    # refleksi
    reflected = nonlocal_square.copy()
    reflected[:,1] = -reflected[:,1]

    # tutup bentuk
    sq = np.vstack([nonlocal_square, nonlocal_square[0]])
    ref = np.vstack([reflected, reflected[0]])

    # update plot
    original_plot.set_data(sq[:,0], sq[:,1])
    reflection_plot.set_data(ref[:,0], ref[:,1])

    # hapus label lama
    for t in texts:
        t.remove()
    texts.clear()

    # label titik
    labels = ['A','B','C','D']
    for i, label in enumerate(labels):
        x, y = nonlocal_square[i]
        xr, yr = reflected[i]

        texts.append(ax.text(x+0.1, y+0.1, f"{label}({x},{y})",
                             color='blue', fontsize=9))

        texts.append(ax.text(x+0.1, yr-0.8, f"{label}'({xr},{yr})",
                             color='red', fontsize=9))

    return original_plot, reflection_plot, *texts

ani = FuncAnimation(fig, update, frames=steps, interval=800)

HTML(ani.to_jshtml())
```
